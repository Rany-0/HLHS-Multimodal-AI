```python
# 基于机器学习模型预测结果的单变量和多变量逻辑回归分析
# 回归模型的OR（Odd Ratios）值和置信区间计算
import pandas as pd
import numpy as np
from scipy import stats
train_df = pd.read_csv(r"D:\wxx\SFYdataJPG\Code\imagescore5\imagescore5\vae\vae0326\train_normalize.csv")
# test_df = pd.read_excel(r"E:\SFYdataJPG\Code\data0103\test-clinial.xlsx")
# # 选择要归一化的列，
# column_to_normalize = [ 'age','fetal_clinical_gestational_age','fetal_ultrasound_gestational_age',	'LA_diameter',
#                     'RA_diameter','LV_diameter','RV_diameter','LA_area','LV_area',
#                     'RA_area','RV_area','LA_RA_area_ratio','LV_RV_area_ratio','left_ventricle_right_ventricle',	' LV_ellipticity',' RV_ellipticity','Image_Score']#2
# column_to_normalize = [ 'age','pregnant_week',	'LA_diameter',
#                     'RA_diameter','LV_diameter','RV_diameter','LA_area','LV_area',
#                     'RA_area','RV_area','LA_RA_area_ratio','LV_RV_area_ratio','left_ventricle_right_ventricle',	' LV_ellipticity',' RV_ellipticity','Image_Score']#2
# 计算Min-Max归一化
# for column in column_to_normalize:

#     train_df[column] = (train_df[column] - train_df[column].min()) / (train_df[column].max() - train_df[column].min())
#     test_df[column] = (test_df[column] - test_df[column].min()) / (test_df[column].max() - test_df[column].min())
#进行z-score归一化
# train_df[column_to_normalize] = stats.zscore(train_df[column_to_normalize])
# test_df[column_to_normalize] = stats.zscore(test_df[column_to_normalize])
# print(train_df)
# ##选择训练集还是全集
train_df = pd.concat([train_df],axis=0)
# #所选择的特征
# print(train_df.columns)

# selected_columns = [ 'Age','PI_RADS','LS', 'MS',  'PSA']#1
# selected_columns = [ 'age','fetal_clinical_gestational_age','fetal_ultrasound_gestational_age',	'LA_diameter',
#                     'RA_diameter','LV_diameter','RV_diameter','LA_area','LV_area',
#                     'RA_area','RV_area','LA_RA_area_ratio','LV_RV_area_ratio','left_ventricle_right_ventricle',	' LV_ellipticity',' RV_ellipticity','Image_Score']#2
# selected_columns = [ 'age','gestational_week',	'LA_diameter',
#                     'RA_diameter','LV_diameter','RV_diameter','LA_area','LV_area',
#                     'RA_area','RV_area','LA_RA_area_ratio','LV_RV_area_ratio','interventricular_septal_thickness',	' LV_ellipticity',' RV_ellipticity','Image_Score']#单
selected_columns = [ 'age','gestational_week',	'LA_diameter',
                    'LV_diameter','LA_area','LV_area',
                    'RV_area','LA_RA_area_ratio','LV_RV_area_ratio',' LV_ellipticity',' RV_ellipticity','Image_Score']#多
# selected_columns = ['pregnant_week','LA_diameter',
#                     'LV_diameter','LA_area','LV_area',
#                     'LA_RA_area_ratio','LV_RV_area_ratio','left_ventricle_right_ventricle',' LV_ellipticity','Image_Score']#2
# selected_columns = [ 'Age','PI_RADS','LS', 'MS',  'PSAD']#3
# selected_columns = [ 'Age','PI_RADS','PSA' ]#4
# selected_columns = [ 'Age','PI_RADS','PSAD']#5


X = train_df[selected_columns]
Y = train_df.iloc[:,1]
# print(X)
# print(Y) #y是label,x是特征
import statsmodels.api as sm
def reg_OR_CI(model):
    """
    import pandas as pd
    import numpy as np

    model - statsmodels.api模型对象
    """

    tmp = pd.DataFrame(
        {
            "OR": model.params,
            "Lower CI": model.conf_int()[0],
            "Upper CI": model.conf_int()[1],
        }
    )

    return np.exp(tmp).iloc[1:,:].values

# 单变量线性回归分析
# ULR_inex = []
# ULR_OR = []
# ULR_CI1 = []
# ULR_CI2 = []
# ULR_pvalue = []

# index = []
# print("[ML CLASS] Variable (Univariate Linear Regression) OR (95% CI) p-value ")
# for i in range(X.shape[1]):
#     X_new = X.iloc[:,i]
#     X_new = sm.add_constant(X_new)
#     linReg = sm.OLS(Y,X_new).fit()
#     OR, CI1, CI2 = reg_OR_CI(linReg)[0]
    
#     print("{:25s} {:.3f} {:.3f} {:.3f} {:.3f}".format(linReg.pvalues.index[1],OR,CI1,CI2,linReg.pvalues[1]))
#     ULR_inex.append(linReg.pvalues.index[1])
#     ULR_OR.append(OR)
#     ULR_CI1.append(CI1)
#     ULR_CI2.append(CI2)
#     ULR_pvalue.append(linReg.pvalues[1])
#     # 保留p<0.05的特征变量
#     if linReg.pvalues[1]<0.05:
#         index.append(i)

#多变量线性回归分析
MLR_inex = []
MLR_OR = []
MLR_CI1 = []
MLR_CI2 = []
MLR_pvalue = []

print("\n多变量回归分析\n")
print("[ML CLASS]Variable (Multivariate Linear Regression) OR (95% CI) p-value ")
X_new = sm.add_constant(X)
linReg = sm.OLS(Y,X_new).fit()
tmp = reg_OR_CI(linReg)
sFeatures = []
ind = []

for i in range(X.shape[1]):
    OR, CI1, CI2 = tmp[i]
    MLR_inex.append(linReg.pvalues.index[i+1])
    MLR_OR.append(OR)
    MLR_CI1.append(CI1)
    MLR_CI2.append(CI2)
    MLR_pvalue.append(linReg.pvalues[i+1])
    if linReg.pvalues[i+1]<0.05:
        print("{:25s} {:.3f} {:.3f} {:.3f} {:.3f}".format(linReg.pvalues.index[i+1],OR,CI1,CI2,linReg.pvalues[i+1]))
        sFeatures.append(linReg.pvalues.index[i+1])
        ind.append(i)
    else:
        print("【删除】{:25s} {:.3f} {:.3f} {:.3f} {:.3f}".format(linReg.pvalues.index[i+1],OR,CI1,CI2,linReg.pvalues[i+1]))
print("保留{:2d} 个特征用于绘制诺莫图！".format(len(sFeatures)))
print()
```

```python
data = {
    'featureName': ULR_inex,
    'ULR_OR': ULR_OR,
    'ULR_CI1': ULR_CI1,
    'ULR_CI2': ULR_CI2,
    'ULR_pvalue': ULR_pvalue,
    # 'featureName': MLR_inex,
    # 'MLR_OR': MLR_OR,
    # 'MLR_CI1': MLR_CI1,
    # 'MLR_CI2': MLR_CI2,
    # 'MLR_pvalue': MLR_pvalue
}
# 创建DataFrame
df = pd.DataFrame(data)
# # 设置小数位数

df['ULR_OR'] = df['ULR_OR'].round(3)
df['ULR_CI1'] = df['ULR_CI1'].round(3)
df['ULR_CI2'] = df['ULR_CI2'].round(3)
df['ULR_pvalue'] = df['ULR_pvalue'].round(4)
# df['MLR_OR'] = df['MLR_OR'].round(3)
# df['MLR_CI1'] = df['MLR_CI1'].round(3)
# df['MLR_CI2'] = df['MLR_CI2'].round(3)
# df['MLR_pvalue'] = df['MLR_pvalue'].round(4)

# 创建新列，格式为"列一 ± 列二"
df['ULR_OR(95%CI)'] = df['ULR_OR'].astype(str) + '' +'('+ df['ULR_CI1'].astype(str)+'-'+df['ULR_CI2'].astype(str)+')'
# df['MLR_OR(95%CI)'] = df['MLR_OR'].astype(str) + '' +'('+ df['MLR_CI1'].astype(str)+'-'+df['MLR_CI2'].astype(str)+')'

# 删除旧的两列
df = df.drop(['ULR_OR', 'ULR_CI1','ULR_CI2'], axis=1)
# df = df.drop(['MLR_OR', 'MLR_CI1','MLR_CI2'], axis=1)

# 重新排列列的顺序
# df = df[['featureName','MLR_OR(95%CI)','MLR_pvalue']]
df = df[['featureName','ULR_OR(95%CI)','ULR_pvalue']]
# # 打印DataFrame
print(df)
df.to_excel(r'E:\SFYdataJPG\Code\imagescore5\imagescore5\vae\vae0326\result\{}_clinic_table单变量0326归一化.xlsx'.format(2),index=False)
```
